## Tratamiento de datos

```python
# Archivo docx a txt

MY_TEXT = docx2txt.process(nombre_archivo+".docx")

with open(nombre_archivo+".txt", "w") as text_file:
    print(MY_TEXT, file=text_file)
    # print(text_file.encoding)

# #---------------

df2 = pd.read_csv(nombre_archivo+".txt", sep = "\r\n", encoding='utf-8', names=['texto'], index_col = False) #'cp1252'

df=pd.DataFrame()
dicc = {}
for linea in df2.index:
    linea = df2[df2.columns[0]][linea]
    if '@' in linea:
        tema = linea

    # if ':' in linea:
    #     datos = linea.split(':')
    if ';' in linea:
        datos = linea.split(';')
        dicc['Tema'] = tema[1:]
        dicc['Sujeto'] = datos[0]
        dicc['Texto'] = datos[1]
        dicc['Filtro'] = filtro
        dicc['Segmento1'] = segmento1
        dicc['Segmento2'] = segmento2
        dicc['Fecha'] = fecha
        dicc['N° estudio'] = n_estudio
        dicc['N° grupo'] = n_grupo

        df=df.append(dicc, ignore_index=True)

# df
# df.drop(axis=1)
# df.reset_index(drop=True, inplace=True)
df.to_excel(nombre_archivo+'.xlsx', index = False)
```

## Líneas de concordancia Docx

```python
# Leer excel y pasarlo a csv

df = pd.read_excel(nombre_archivo + '.xlsx')

# Eliminar espacios al inicio y final
for i in df["Texto"]:
    df["Texto"] = df["Texto"].replace(i, str(i).strip())

# Eliminar moderador y moderadora
df_aux = pd.DataFrame(df, columns= ['Sujeto', 'Texto'])
indexNames = df_aux[ (df_aux['Sujeto'] == 'Moderador')].index 
indexNames1 = df_aux[(df_aux['Sujeto'] == 'Moderadora')].index
indexNames = indexNames.append(indexNames1)
df_aux.drop(indexNames , inplace=True)

# eliminar años y comuna
lista_aux = (df_aux['Sujeto'].str.split('('))
lista_sujeto = []
for i in lista_aux:
    lista_sujeto.append(i[0])
df_aux['Sujeto'] = lista_sujeto
df_aux['Texto_full'] = df_aux['Sujeto'] + ': '+ df_aux['Texto']
df_aux
```

```python
# Buscar lista de palabras y generar archivo docx y xlsx

document = docx.Document()
archivo = 'lineas de concordancia2.txt'
df_lineas = pd.DataFrame(columns=['sujeto', 'texto antes', 'palabra', 'texto después'])

resultado = []
for i in lista:
    # conc_results4 = ct.concord(ct.tokenize(ct.ldcorpus("brown_single"),lemma = False),["run.*","ran"], nhits = 10, regex = True)
    conc_results1 = ct.concord(ct.tokenize(df_aux['Texto_full'], lemma=False), [i], nhits = 100, nleft=2000, nright=2000, regex = True, sep=',')#, outname='lineas de concordancia DF.txt')
    if bool(conc_results1) == False:
        pass
    else:
        resultado.extend(conc_results1)

for linea2 in resultado:
    izq = ' '.join(linea2[0])
    palabra = linea2[1]
    der =' '.join(linea2[2])

    # escribir = str((izq +' '+ palabra+' '+ der+ '\n\n'))#.encode('utf8'))
    # xlsx------------------------------
    if linea2[0][0] == 'mujer' or linea2[0][0] == 'hombre':
        sujeto = linea2[0][0] + ' ' + linea2[0][1]
        print(sujeto)
    izq_2 = ' '.join(linea2[0][2:])
    df_lineas=df_lineas.append({'sujeto':sujeto, 'texto antes' : izq_2 , 'palabra' : palabra, 'texto después' : der} , ignore_index=True)
    df_lineas.to_excel('lineas_concordancia - ' + nombre_archivo + '.xlsx', index=False)
    # docx-----------------------------
    p = document.add_paragraph(izq + ' ')
    p.add_run(palabra).font.color.rgb = RGBColor(0,0,255)
    p.add_run(' ' + der)
    document.save('lineas_concordancia - ' + nombre_archivo + '.docx')
    # -----------------------------
```
